# Eliminación de Outliers mediante el rango intercuartílico en la columna SalePrice

```python
data_num["SalePrice"].plot(kind='box')
```

```python
data_num.describe().loc["25%", "SalePrice"]
```

```python
# Calcular el rango intercuartílico (IQR)
Q1 = data_num["SalePrice"].quantile(0.25)
Q3 = data_num["SalePrice"].quantile(0.75)
IQR = Q3 - Q1
print(IQR, Q1, Q3)
```

```python
# Definir los límites para identificar outliers
limite_inferior = Q1 - 2 * IQR
limite_superior = Q3 + 1.5 * IQR

# Filtrar los outliers
outliers = data_num[(data_num["SalePrice"] < limite_inferior) | (data_num["SalePrice"] > limite_superior)]
no_outliers = data_num[~((data_num["SalePrice"] < limite_inferior) | (data_num["SalePrice"] > limite_superior))]

# Imprimir los resultados
print("Límite inferior:", limite_inferior)
print("Límite superior:", limite_superior)
print("Número de outliers:", len(outliers))
print("Outliers:")
print(outliers)
```

```python
no_outliers["SalePrice"].plot(kind='box')
```

Puedo concatenar los datos que no tienen outliers con los datos que sí los tienen, y a los cuales les he eliminado la columna SalePrice.

De esa manera, las casillas de la columna que tenían valores excesivos, están ocupadas por "NaN"

```python
pd.concat([outliers.drop(columns=["SalePrice"]), no_outliers], sort=True)["SalePrice"]
```

También puedo quedarme con el índice de los outliers...

```python
outliers.index
```

...y usar ese índice para ubicar en las casillas

```python
data["SalePrice"].loc[outliers.index] = limite_superior
```

Comprobamos la modificación

```python
data.iloc[50:60,:]
```

# Eliminación de Outliers mediante la desviación típica en cualquier columna usando una función

```python
def limpia_columna(data_num, columna, veces):
    # Calcular la media y la desviación típica
    media = data_num[columna].mean()
    desviacion_tipica = data_num[columna].std()

    # Definir los límites para identificar outliers
    limite_inferior = media - veces * desviacion_tipica
    limite_superior = media + veces * desviacion_tipica

    # Filtrar los outliers
    outliers = data_num[(data_num[columna] < limite_inferior) | (data_num[columna] > limite_superior)]
    print("Índices de los outliers:", outliers.index)
    no_outliers = data_num[~((data_num[columna] < limite_inferior) | (data_num[columna] > limite_superior))]

    # Imprimir los resultados
    print("Límite inferior:", limite_inferior)
    print("Límite superior:", limite_superior)
    print("Número de outliers:", len(outliers))

    data[columna].loc[outliers.index] = limite_superior
    return data

```

```python
datos_limpios = limpia_columna(data_num, "LotArea", 3)
```
